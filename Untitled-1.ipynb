### Data Credibility Analysis

I will be implementing the **ROCC** method to determine the credibility of the data and inform us of any bias present in the data.

* Reliability: 
  * Result: Not Reliable
    * Reason: The sample size in the dataset is too small. There are a total of 30 users that are tracked in the dataset, this number is not large enough
      to eliminate sample bias in a population of 30 million FitBit users. With the given 30 user sample size out of a total population of over 30 million
      this would give us a margin of error of 17.5%, which is not enough. To achieve a desired 5% margin of error with a 95% confidence level you would need
      a test population of ~385 users. This amount of 30 users does still meet the minimum required sample size for Central Limit Theorem (CLT). Although,
      the data was only collected over a two month period, which is not long enough to give a high degree of confidence. Given the option I would prefer to
      have 6 months to 1 year of data to be more accurate.
* Originality:
  * Result: Data is not Original.
    * Reason: The data was collected via a survey conducted by a 3rd party company (Amazon Mechanical Turk). First party information collected directly from 
              Fitbit would be the preferred resource.
* Comprehensive:
  * Result: Data is not Comprehensive.
    * Reason: The data should not be considered comprehensive, because there is relevant related data that is not provided that would be useful in making 
              our analysis more complete. Some examples of this missing data would be: sex, age, height, etc. Also, as I stated above, there is the issue
              of a minimally sized sample group and an less than ideal amount of time over which the data was collected. I would also like to point out
              that we were provided no information on how the sample group participants were chose, thus we cannot rule out bias. If we were able to prove
              the users were selected at random this would be better as we could rule out bias in the data.
* Current:
  * Result: Data is not current
    * Reason: The data was collected over 6 years ago in 2016.
* Cited:
  * Result: The data IS cited but NOT Credible.
    * Reason: We know the data comes from Amazon Mechanical Turk (it is cited as such). But not much is known about this source and more research would be required
              to prove or disprove its credibility. However, this is out of the scope of the business ask.

To summarize, the data from this analysis should be considered with a healthy degree of skepticism and caution. Despite the above listed shortcomings in the data it
is still possible to come away with useful insights to future trends in the market and valuable information to the business.

```r
activity_new_v2 <- activity_daily %>% 
  filter(TotalSteps !=0)

View(activity_new_v2)
```

After removing the zero activity days I am left with some suspiciously low activity days, however, I can not remove these because there is a chance </br>
that the patient is restricted to a bed all day. I have also noticed some very low calorie burn entries for users on certain days and will leave these </br>

There are also entries missing for some users sleep or weight. I will assume that users forgot on those days and will not make any changes to account for this.

Overall the dataset is far from perfect but so far we are making the most of what we can from it.

The next issue I have found is the date and time formatting. Both date and time are stored in the same column. This could be fine if I didn't think I would need</br>
these fields for any analysis between the three files, however, I can not confidently decide this so I will split the field into two (Date, Time)



```r
weight_new_v2 <- weight %>% 
  separate(Date, c("Date", "Time"), " ")

sleep_daily_v2 <- sleep_daily %>% 
  separate(SleepDay, c("Date", "Time"), " ")
```

In light of the lower confidence level with the data I will be checking for duplication of data as well. We already had missing data, not enough data, and too much data,</br>

## Time for Visuals!

Because not many people enjoy just staring at black and white numbers.

```r
VeryActiveMin <- sum(activity_daily_final$VeryActiveMinutes)
FairlyActiveMin <- sum(activity_daily_final$FairlyActiveMinutes)
LightlyActiveMin <- sum(activity_daily_final$LightlyActiveMinutes)
SedentaryMin <- sum(activity_daily_final$SedentaryMinutes)
TotalMin <- VeryActiveMin + FairlyActiveMin + LightlyActiveMin + SedentaryMin

slices <- c(VeryActiveMin,FairlyActiveMin,LightlyActiveMin,SedentaryMin)
lbls <- c("VeryActive","FairlyActive","LightlyActive","Sedentary")
pct <- round(slices/sum(slices)*100)
lbls <- paste(lbls, pct)
lbls <- paste(lbls, "%", sep="")
pie(slices, labels = lbls, col = rainbow(length(lbls)), main = "Percentage of Activity in Minutes")

```

Here we have a simple pie chart showing the percentage of time the average user spent at each level of activity. </br>
As you can see most of their time is spent sedentary which leave a lot of room for improvement in the amount of time</br>
that users spend active in order to achieve their fitness goals. Bellabeat can capitalize on this subset of the population </br>
that could greatly benefit from more active time in a day. 

```r
ggplot(data=activity_daily_final) +
  geom_point(mapping=aes(x=TotalSteps, y=Calories), color="red") +
  geom_smooth(mapping=aes(x=TotalSteps, y=Calories)) +
  labs(title="The Relationship Between Total Steps and Calories Burned", x="Total Steps", y="Calories Burned (kcal)")
```

So this graph shows the relationship between steps taken and calories burned. This relationship is kind of an obvious one, if a person walks more they generally burn more calories, and if they burn more calories odds are they walk more. Much more insight is hard to tell unless we can get data on sex, height, accurate weight, age, diet, etc.

This does show that there is still a need in the market for assistance in getting users to take more steps, burn more calories, and keep their BMI in a health range.

```r
combined_data <- merge(activity_daily_final, sleep_daily_final, by="Id")

ggplot(data=combined_data) +
    geom_point(mapping=aes(x=TotalMinutesAsleep, y=VeryActiveMinutes, color="VeryActiveMinutes")) +
    geom_smooth(mapping=aes(x=TotalMinutesAsleep, y=VeryActiveMinutes, regLineColor="blue"))+
    labs(title="The Relationship Between Activity Levels and Total Minutes Asleep", x="Total Minutes Asleep", y="Minutes of Activity")
```

```r
ggplot(data=combined_data) +
    geom_point(mapping=aes(x=TotalMinutesAsleep, y=LightlyActiveMinutes, color="LightlyActiveMinutes")) +
    geom_smooth(mapping=aes(x=TotalMinutesAsleep, y=LightlyActiveMinutes, regLineColor="blue")) +
    labs(title="The Relationship Between Activity Levels and Total Minutes Asleep", x="Total Minutes Asleep", y="Minutes of Activity")
```

```r
ggplot(data=combined_data) +
    geom_point(mapping=aes(x=TotalMinutesAsleep, y=SedentaryMinutes, color="SedentaryMinutes")) +
    geom_smooth(mapping=aes(x=TotalMinutesAsleep, y=SedentaryMinutes, regLineColor="blue")) +
    labs(title="The Relationship Between Activity Levels and Total Minutes Asleep", x="Total Minutes Asleep", y="Minutes of Activity")
```

I decided to try and map out the relationship between hours of sleep and each activity level. I had speculated that the longer a </br> 
person slept the more active they could be, however the data proves this is not accurate and there are cases in which users who slept</br> 
longer were more sedentary. This insight makes sense since the longer you sleep the less of a window you have left in a day to be active. </br>
There is a balance in all of this where if you don't get enough sleep you are more sedentary and if you get to much you are more sedentary as well.
